```python
inventory = {
    "t-shirt": 5,
    "mug": 3,
    "hat": 2,
    "book": 4,
    "keychain": 1
}
```

```python
customer_orders = {"mug", "hat", "pen"}
```

```python
for product in customer_orders:
    if product in inventory and inventory[product] > 0:
        inventory[product] -= 1
```
